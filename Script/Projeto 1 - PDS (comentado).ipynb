```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#PARTE 1

#Função para gerar o vetor de instantes (n) e o vetor com amostras de x(t), de frequência o0, em cada instante (xn).
#Para uma frequencia de amostragem constante de fs = 8191 Hz.
def generator(o0):
    oS = 2*np.pi*8192

    n = np.arange(8192)
    xn = np.sin(o0*n/8192)
    
    return n, xn
```

```python
#Plotando os gráficos de tempo contínuo e de tempo discreto para as 50 primeiras amostras de x(t)
def plotting(n,xn):
    plt.subplot(2,1,1)
    plt.plot(n[:50],xn[:50])
    plt.grid()
    plt.xlabel('t (tempo)')
    plt.ylabel('x(t)')

    plt.subplot(2,1,2)
    plt.stem(n[:50],xn[:50])
    plt.grid()
    plt.xlabel('t (tempo)')
    plt.ylabel('x[n]')
    return
```

```python
def ctfts(x,T):
    N = len(x)
    X = np.fft.fftshift(np.fft.fft(x,N)*(2*np.pi/N))
    
    index1 = np.where(abs(X.real)<1)
    X.real[index1] = 0

    index2 = np.where(abs(X.imag)<1)
    X.imag[index2] = 0
    
    o = np.angle(X)
    A = np.abs(X)
    w = np.linspace(-1,1-1/N,N)/(2*T)
    
    #Plot de Amplitude e Fase da Transformada
    plt.subplot(2,1,1)
    plt.plot(w, A)
    plt.grid()
    plt.xlabel('w (frequência)')
    plt.ylabel('|X(e^jw)| (Amplitude)')

    plt.subplot(2,1,2)
    plt.plot(w, o)
    plt.grid()
    plt.xlabel('w (frequência)')
    plt.ylabel('<(X(e^jw)) (fase)')

    return A,o,w
```

```python
#Gerando as amostras de x(t) para fo
n, xn = generator(2*np.pi*1000)

#Plotando x(t) e xn
plotting(n,xn)
```

```python
#Transformada de Fourier de Tempo Continuo
mag,fase,w = ctfts(xn,1/8192)
```
